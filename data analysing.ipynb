```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.titlesize'] = 16
```

```python
emp_df = pd.read_excel('employee.xlsx')
std_df = pd.read_excel('std_marks_sheet.xlsx')
```

```python
emp_df
```

```python
# PART 1: EMPLOYEE DATASET INSIGHTS 
```

```python
# Insight 1: Average Salary by Designation
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=emp_df, x='Designation', y='Salary', estimator='mean', errorbar=None, palette='viridis')
plt.title('Insight: Average Salary Distribution by Designation')
plt.xlabel('Designation')
plt.ylabel('Average Salary ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Insight 2: Departmental Workforce Distribution
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=emp_df, x='Department', palette='magma', order=emp_df['Department'].value_counts().index)
plt.title('Insight: Number of Employees per Department')
plt.xlabel('Department')
plt.ylabel('Employee Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Insight 3: Age vs Salary Relationship
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=emp_df, x='Age', y='Salary', hue='Designation', s=100, palette='deep')
plt.title('Insight: Relationship between Age and Salary')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python

# PART 2: STUDENT DATASET INSIGHTS
```

```python
subjects = ['python', 'numpy', 'pandas', 'streamlit', 'ML', 'DL', 'NLP']
```

```python
# Insight 1: Comparison of Subject Scores (Boxplot)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=std_df[subjects], palette='Set2')
plt.title('Insight: Spread and Outliers of Marks across Subjects')
plt.ylabel('Marks Obtained')
plt.xlabel('Subject Name')
plt.show()
```

```python
# Insight 2: Grades distrunuted by (Pie Chart)
```

```python
plt.figure(figsize=(8, 8))
grade_counts = std_df['Grade'].value_counts()
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', 
        colors=sns.color_palette('pastel'), startangle=140, explode=[0.05] * len(grade_counts))
plt.title('Insight: Proportion of Student Grades')
plt.show()
```

```python
# Insight 3: Subject Inter-dependency throught (Heatmap)
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = std_df[subjects].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Insight: Correlation Heatmap of Subject Performance')
plt.show()
```

```python
# Insight 4: Distribution of Total Marks
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(std_df['Total'], kde=True, color='skyblue', bins=15)
plt.title('Insight: Distribution of Total Marks Scored')
plt.xlabel('Total Marks')
plt.ylabel('Frequency')
plt.show()
```
